```python
import pandas as pd

df = pd.read_csv("rfm_modeling_snapshot.csv")

df.head()
df.columns
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,f1_score,roc_auc_score
import joblib

# features

features=[

'recency_days',

'frequency_180d',

'monetary_180d',

'return_rate_180d',

'avg_discount_pct_180d',

'avg_rating_180d',

'category_diversity_180d',

'sessions_30d',

'product_views_30d',

'cart_adds_30d',

'wishlist_adds_30d',

'abandoned_carts_30d',

'email_opens_30d',

'campaign_clicks_30d',

'last_visit_days_ago'

]

# split

train=df[df['split']=='train']

test=df[df['split']=='test']

# model

model=RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

model.fit(

train[features],

train['churn_next_60d']

)

# prediction

pred=model.predict(

test[features]

)

prob=model.predict_proba(

test[features]

)[:,1]

print("Accuracy")

print(

accuracy_score(

test['churn_next_60d'],

pred

)

)

print("\nF1")

print(

f1_score(

test['churn_next_60d'],

pred

)

)

print("\nROC AUC")

print(

roc_auc_score(

test['churn_next_60d'],

prob

)

)

joblib.dump(

model,

'model.pkl'

)

print("\nmodel saved")
```
